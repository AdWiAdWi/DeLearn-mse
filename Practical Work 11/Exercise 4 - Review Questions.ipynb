__a)__ How can you define a generative system ? Describe the two approaches seen in the class
to build generative systems with RNNs.
- - -
A generative system is a system able to generate new consistent data from a seed. By consistent, we mean respecting temporal and spatial "structures" that have been learned from the input space.

We can distinguish two approaches approaches (__Many to one__ and  __Many to many__). The principle is to train the RNN to predict the next token of the sequence in a many to one setting. The difference for many to many is mainly at training time where the RNN cells are emitting an output at each step.

__c)__ Give two desired properties of word embeddings when used in a classification task (such as sentiment classification).
- - -
- In the embedding space, similar words should be close to each other and dissimilar ones far apart such that a classifier can easily deduce a meaningful decision boundary. Hence we want the space to exploit "good" features.
- The embedding space should already be initialized with basic knowledge (pretraining) and should only be fine tuned on a specific task such as sentiment classification.

__g)__ Draw the architecture of a Seq2Seq model and explain the encoder/decoder concept.
- - -
The Seq2Seq architecture consists of two main building blocks, an encoder and a decoder. The encoder first reads the whole input sentence and somehow encodes the knowledge contained in it which could be refered to as a context based embedding. After the encoder is finished, the decoder starts to generate the required output sentence, based on the knowledge previously encoded by the encoder and on the generated token by the decoder itself.


__h)__ What is the main problem of a basic Seq2Seq model used for machine translation and the solution that can be used to overcome this problem?
- - -
The mapping for short sequences is working well but the mapping for longer sequences is more difficult. The context vectors needs to "encode" a lot of information. Simply put, it can not remember the beginning of the input sentence if it is too long.

The solution to this are __Seq2Seq models with attention__. The attention helps to remind the model on which part it should focus. Another more advanced solution are __Transformers__.
