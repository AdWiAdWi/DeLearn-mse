__a)__ What is the difference between a Dense and a Conv2D layer in Keras ?
- - -
A __Dense__ layer is a fully connected layer. It can be viewed as a convolution layer with K filters spanning the
full input space. It is often used at the output of the CNN architecture.

In a __convolutional__ layer a neuron is only connected to a specified, local set of neurons of the previous layer. In the Conv2D layer this set is defined by a two dimensional grid which is parametrized by the kernel size.

__b)__ In CNN, explain what it means to “preserve the spatial structure” ?
- - -
In dense layers, we completely loose the information on the spatial structure as the output neurons are fully connected.
E.g. we could shuffle the pixels randomly (but equally for all images) and the network would perform the same. In a CNN the spatial structure is preserved because each neuron activation depends only on a receptive field in the previous layer.

__c)__ In CNN, explain the different operations taking place when performing a convolution of a
filter on an input, e.g. an image. How is the filter defined in terms of hyper-parameters?
- - -
The convolution is parametrized by the __kernel size__ (spatial size of the local sliding window), __stride__ (amout of pixels the window is shifted after each step) and the __padding__ (amount of pixels of value 0 added on the border). For a convolution on an image with  channels using a $3 \times 3$ kernel we have $3 \times (3 \times 3) + 1$ weights. The result of the convolution for a given window is just the dot product of the kernel $w$ with the window values $v$ and adding the bias term $b$.


$logit = w \cdot v + b$



__d)__ Define what is meant by receptive field and activation map.
- - -
The sub-area of an input that influences a component of the output is called the __receptive field__ of the latter.

A __activation map__ is the result of the convolution of the filter, i.e. the sliding over all possible spatial locations of the image.

__e)__ Explain the effect of a given stride or zero-padding value.
- - -
The __stride__ defines the amount of pixels the filter(local window) is shifted after each step. A stride bigger than 1 will lead to significantly smaller output.


By adding __zero-padding__ we can prevent that the output is shrinked too much or is even the same as the input. Therefore, zeros are added on the border of the input.

__g)__ Define what is a max pooling layer and what is its role in a CNN architecture.
- - -
When a max pooling layer is applied the spatial size of the representation is reduced by only keeping the max value within a spatially defined local window. The most significant activations are kept and it reduces the amount of computation and also controls overfitting.
